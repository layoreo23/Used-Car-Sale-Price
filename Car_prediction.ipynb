```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('car data.csv')
```

```python
final=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
final['Current_Year']=2023
```

```python
final.head()
```

```python
final['No_Years']=final['Current_Year']-final['Year']
final.head()
```

```python
final.drop(['Year'],axis=1,inplace=True)
```

```python
final.drop(['Current_Year'],axis=1,inplace=True)
```

```python

final.head() #final data set
```

```python
final=pd.get_dummies(final,drop_first=True)
```

```python
final
```

```python
cor=final.corr()
cor
```

```python
sns.pairplot(final)
```

```python
top_features=cor.index
plt.figure(figsize=(20,20))
sns.heatmap(final[top_features].corr(),annot=True)
```

```python
X=final.iloc[:,1:] #independent 
y=final.iloc[:,0]  #dependent
```

```python
pip install scikit-learn
```

```python
#sorting by feature importance
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
feat_imp=pd.Series(model.feature_importances_,index=X.columns)
feat_imp.nlargest(4).plot(kind='barh')
plt.show()
```

```python
#training
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
```

```python
#hyperparameters
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=100,stop=1200,num=12)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=6, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
pred=rf_random.predict(X_test)
pred
```

```python
sns.distplot(y_test-pred)
```

```python
plt.scatter(y_test,pred) #linear ~kind of better
```

```python
import pickle
file=open('random_dorest_regression.pkl','wb')
pickle.dump(rf_random,file)
```
